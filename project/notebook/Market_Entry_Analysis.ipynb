```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Load data
file_path = r'D:\personal\bdas.spyberpolymath.com\project\data\Market_Entry.xlsx'
df = pd.read_excel(file_path)
```

```python
# Display first rows
print(df.head())
```

```python
# Basic statistics
print("\nData Summary:")
print(df.describe())
```

```python
# Visualize Competitor Count vs Potential Customers
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='CompetitorCount', y='PotentialCustomers', hue='Region', s=100)
plt.title("Competitor Count vs Potential Customers by Region")
plt.xlabel("Competitor Count")
plt.ylabel("Potential Customers")
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
# KMeans clustering to segment markets
X = df[['CompetitorCount', 'PotentialCustomers']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='CompetitorCount', y='PotentialCustomers', hue='Cluster', palette='Set1', s=100)
plt.title("Market Segmentation using KMeans Clustering")
plt.xlabel("Competitor Count")
plt.ylabel("Potential Customers")
plt.grid(True)
plt.show()
```

```python
# Save clustered data
df.to_excel(r'D:\personal\bdas.spyberpolymath.com\project\data_output\Market_Entry_Clustered.xlsx', index=False)
```
