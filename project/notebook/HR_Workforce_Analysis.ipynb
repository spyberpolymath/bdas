```python
# HR Workforce Planning Analysis
# Author: Aman Anil
# Date: 2025-10-12

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file
file_path = r"D:\personal\bdas.spyberpolymath.com\project\data\HR_Workforce_Planning.xlsx"
df = pd.read_excel(file_path)

# Display first 10 rows
df.head(10)
```

```python
# Summary statistics
df.describe()

# Check unique skills
print("Unique Skills:", df['Skill'].unique())

# Total hours worked by Skill
hours_by_skill = df.groupby('Skill')['HoursWorked'].sum().sort_values(ascending=False)
print(hours_by_skill)

# Plot total hours worked by Skill
plt.figure(figsize=(8,5))
sns.barplot(x=hours_by_skill.index, y=hours_by_skill.values, palette="viridis")
plt.title("Total Hours Worked by Skill")
plt.ylabel("Hours Worked")
plt.xlabel("Skill")
plt.show()

# Total hours worked by Project
hours_by_project = df.groupby('ProjectAllocated')['HoursWorked'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=hours_by_project.index, y=hours_by_project.values, palette="magma")
plt.title("Total Hours Worked by Project")
plt.ylabel("Hours Worked")
plt.xlabel("Project")
plt.show()
```
