```python
# Supply Chain Optimization Simulation
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.optimize import linprog
import matplotlib.pyplot as plt
```

```python
# 1️⃣ Load data directly from your Excel file
file_path = r"D:\personal\bdas.spyberpolymath.com\project\data\Supply_Chain_Optimization.xlsx"
df = pd.read_excel(file_path)
```

```python
print("Original Data:")
display(df)
```

```python
# 2️⃣ Reorder Analysis - check which products need reordering
df['ReorderNeeded'] = df['StockLevel'] < df['ReorderLevel']
print("\nAfter Reorder Analysis:")
display(df)
```

```python
# 3️⃣ Simple Cost Optimization (illustrative example)
# We simulate an "Optimal Order Quantity" as difference between ReorderLevel and StockLevel
df['OptimalOrder'] = np.maximum(df['ReorderLevel'] - df['StockLevel'], 0)

print("\nAfter Optimization Simulation:")
display(df)
```

```python
# 4️⃣ Visualize stock vs reorder
plt.figure(figsize=(8,5))
plt.bar(df['ProductID'], df['StockLevel'], label='Stock Level', alpha=0.7)
plt.bar(df['ProductID'], df['ReorderLevel'], label='Reorder Level', alpha=0.5)
plt.xlabel("Product ID")
plt.ylabel("Quantity")
plt.title("Stock vs Reorder Levels")
plt.legend()
plt.show()
```

```python
# 5️⃣ Save the simulation results
output_file = r"D:\personal\bdas.spyberpolymath.com\project\data_output\simulation_report.xlsx"
df.to_excel(output_file, index=False)
print(f"\nSimulation report saved at: {output_file}")
```
