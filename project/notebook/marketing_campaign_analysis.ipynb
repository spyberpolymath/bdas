```python
# 1Ô∏è‚É£ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# 2Ô∏è‚É£ Load Excel data
file_path = r"D:\personal\bdas.spyberpolymath.com\project\data\Marketing_Campaign.xlsx"
df = pd.read_excel(file_path)
```

```python
# 3Ô∏è‚É£ Display basic info
print("üìä Dataset Overview:")
display(df.head())
print("\nColumns:", df.columns.tolist())
print("\nBasic Stats:\n", df.describe())
```

```python
# 4Ô∏è‚É£ Analyze ROI performance
avg_roi = df['ROI'].mean()
top_channel = df.groupby('Channel')['ROI'].mean().idxmax()
print(f"\n‚úÖ Average ROI: {avg_roi:.2f}")
print(f"üèÜ Top Performing Channel: {top_channel}")
```

```python
# 5Ô∏è‚É£ Visualize Spend vs ROI
plt.figure(figsize=(8,5))
plt.scatter(df['Spend'], df['ROI'], c='blue', edgecolors='k')
plt.title('Marketing Campaign: Spend vs ROI')
plt.xlabel('Spend (in currency units)')
plt.ylabel('ROI')
plt.grid(True)
plt.show()
```

```python
# 6Ô∏è‚É£ Channel-level ROI trend
channel_roi = df.groupby('Channel')['ROI'].mean().sort_values(ascending=False)
channel_roi.plot(kind='bar', color='orange', figsize=(6,4))
plt.title('Average ROI by Marketing Channel')
plt.xlabel('Channel')
plt.ylabel('Average ROI')
plt.show()
```

```python
# Prepare summary
summary = {
    'Average ROI': [avg_roi],
    'Top Channel': [top_channel]
}

# Define output path and ensure folder exists
output_dir = r"D:\personal\bdas.spyberpolymath.com\project\data_output"
os.makedirs(output_dir, exist_ok=True)

# Create DataFrame and export to Excel
summary_df = pd.DataFrame(summary)
output_path = os.path.join(output_dir, "Campaign_Summary.xlsx")
summary_df.to_excel(output_path, index=False)

print(f"\nüìÅ Summary saved successfully at: {output_path}")
```
