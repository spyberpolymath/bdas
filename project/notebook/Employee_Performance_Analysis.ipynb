```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
file_path = r"D:\personal\bdas.spyberpolymath.com\project\data\Employee_Performance.xlsx"
df = pd.read_excel(file_path)
```

```python
# Display first few rows
print("Data preview:")
display(df)
```

```python
# Basic statistics
print("\nSummary statistics:")
display(df.describe())
```

```python
# Department-wise performance
dept_perf = df.groupby('Department')['PerformanceScore'].mean().sort_values(ascending=False)
print("\nDepartment-wise average performance:")
display(dept_perf)
```

```python
# Top performers
top_performers = df.sort_values(by='PerformanceScore', ascending=False).head(5)
print("\nTop performers:")
display(top_performers)
```

```python
# Visualizations
plt.figure(figsize=(10,6))
sns.barplot(x='EmployeeID', y='PerformanceScore', hue='Department', data=df)
plt.title('Employee Performance Scores by Department')
plt.xlabel('Employee ID')
plt.ylabel('Performance Score')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Department', y='PerformanceScore', data=df)
plt.title('Performance Score Distribution by Department')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Attendance', y='PerformanceScore', hue='Department', data=df, s=100)
plt.title('Attendance vs Performance Score')
plt.xlabel('Attendance')
plt.ylabel('Performance Score')
plt.show()
```
